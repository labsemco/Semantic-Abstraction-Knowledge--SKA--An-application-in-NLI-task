```python
import pandas as pd
from scipy.stats import ttest_ind
from scipy.stats import mannwhitneyu
```

```python
ruta="../"
models=["gemma2:2b","gemma2","llama3.2","llama3.1","phi3:medium","phi3"]
corpus=["SICK","Scitail","RTEGLUE","SNLI"]
```

```python
new_data={"model":[],"corpus":[],"cv":[],
          "base_avg":[],"base_std":[],"base_std0":[],
          "SKA_DT_avg":[],"SKA_DT_std":[],"SKA_DT_std0":[],"p_value_SKA_DT":[],
          "SKA_MV_avg":[],"p_value_SKA_MV":[],"SKA_MV_std":[],"SKA_MV_std0":[],
          "SAoT_avg":[],"p_value_SAoT":[],"SAoT_std":[],"SAoT_std0":[],
          "SAoT_FS_avg":[],"p_value_SAoT_FS":[],"SAoT_FS_std":[],"SAoT_FS_std0":[],
          "AoT_avg":[],"p_value_AoT":[],"AoT_std":[],"AoT_std0":[]}
```

```python
for m in models:
    for c in corpus:
        for i in range(5):
            print(m,c,"CV",i+1)
            data_results=pd.read_pickle(ruta+m+"/"+c+"/answers/cv"+str(i+1)+"/results.pickle")
            base=data_results[data_results["group"]=="Base"]["accuracy"]
            base_avg=data_results[data_results["group"]=="Base"]["accuracy"].mean()
            base_std=data_results[data_results["group"]=="Base"]["accuracy"].std()
            base_std0=data_results[data_results["group"]=="Base"]["accuracy"].std(ddof=0)

            gs_dt=data_results[data_results["group"]=="SKA_DT"]["accuracy"]
            gs_dt_avg=data_results[data_results["group"]=="SKA_DT"]["accuracy"].mean()
            gs_dt_std=data_results[data_results["group"]=="SKA_DT"]["accuracy"].std()
            gs_dt_std0=data_results[data_results["group"]=="SKA_DT"]["accuracy"].std(ddof=0)

            vm=data_results[data_results["group"]=="SKA_MV"]["accuracy"]
            vm_avg=data_results[data_results["group"]=="SKA_MV"]["accuracy"].mean()
            vm_std=data_results[data_results["group"]=="SKA_MV"]["accuracy"].std()
            vm_std0=data_results[data_results["group"]=="SKA_MV"]["accuracy"].std(ddof=0)

            saot=data_results[data_results["group"]=="SAoT"]["accuracy"]
            saot_avg=data_results[data_results["group"]=="SAoT"]["accuracy"].mean()
            saot_std=data_results[data_results["group"]=="SAoT"]["accuracy"].std()
            saot_std0=data_results[data_results["group"]=="SAoT"]["accuracy"].std(ddof=0)

            saot_fs=data_results[data_results["group"]=="SAoT_FS"]["accuracy"]
            saot_fs_avg=data_results[data_results["group"]=="SAoT_FS"]["accuracy"].mean()
            saot_fs_std=data_results[data_results["group"]=="SAoT_FS"]["accuracy"].std()
            saot_fs_std0=data_results[data_results["group"]=="SAoT_FS"]["accuracy"].std(ddof=0)

            aot=data_results[data_results["group"]=="AoT"]["accuracy"]
            aot_avg=data_results[data_results["group"]=="AoT"]["accuracy"].mean()
            aot_std=data_results[data_results["group"]=="AoT"]["accuracy"].std()
            aot_std0=data_results[data_results["group"]=="AoT"]["accuracy"].std(ddof=0)

            new_data["cv"].append(i+1)
            new_data["model"].append(m)
            new_data["corpus"].append(c)
            new_data["base_avg"].append(base_avg)
            new_data["base_std"].append(base_std)
            new_data["base_std0"].append(base_std0)
            
            new_data["SKA_DT_avg"].append(gs_dt_avg)   
            new_data["p_value_SKA_DT"].append(mannwhitneyu(base, gs_dt).pvalue)
            new_data["SKA_DT_std"].append(gs_dt_std)
            new_data["SKA_DT_std0"].append(gs_dt_std0)

            new_data["SKA_MV_avg"].append(vm_avg)   
            new_data["p_value_SKA_MV"].append(mannwhitneyu(base, vm).pvalue)
            new_data["SKA_MV_std"].append(vm_std)
            new_data["SKA_MV_std0"].append(vm_std0)

            new_data["SAoT_avg"].append(saot_avg)   
            try:
                new_data["p_value_SAoT"].append(mannwhitneyu(base, saot).pvalue)
            except:
                new_data["p_value_SAoT"].append(0)
            new_data["SAoT_std"].append(saot_std)
            new_data["SAoT_std0"].append(saot_std0)

            new_data["SAoT_FS_avg"].append(saot_fs_avg)   
            try:
                new_data["p_value_SAoT_FS"].append(mannwhitneyu(base, saot_fs).pvalue)
            except:
                new_data["p_value_SAoT_FS"].append(0)
            new_data["SAoT_FS_std"].append(saot_fs_std)
            new_data["SAoT_FS_std0"].append(saot_fs_std0)

            new_data["AoT_avg"].append(aot_avg) 
            try:  
                new_data["p_value_AoT"].append(mannwhitneyu(base, aot).pvalue)
            except:
                new_data["p_value_AoT"].append(0)
            new_data["AoT_std"].append(aot_std)
            new_data["AoT_std0"].append(aot_std0)
            
```

```python
def format_pvalue(p):
    if p < 0.001:
        return "***"
    elif p < 0.01:
        return "**"
    elif p < 0.05:
        return "*"
    else:
        return round(p, 4)
    
tabla_avg_pvalue_grouped= temp.groupby(["model","corpus"]).agg({
    "base_avg":"mean",
    "base_std":"mean",
    #"base_std0":"mean",
    "SKA_DT_avg":"mean",
    "SKA_DT_std":"mean",
    #"DT_std0":"mean",
    "p_value_SKA_DT":"mean",
    
    "SAoT_avg":"mean",
    "SAoT_std":"mean",
    #"aot_std0":"mean",
    "p_value_SAoT":"mean",

    "SAoT_FS_avg":"mean",
    "SAoT_FS_std":"mean",
    #"aot_fs_std0":"mean",
    "p_value_SAoT_FS":"mean",
    
    "AoT_avg":"mean",
    "AoT_std":"mean",
    #"cot_std0":"mean",
    "p_value_AoT":"mean",
    "SKA_MV_avg":"mean",
    "SKA_MV_std":"mean",
    #"vm_std0":"mean",
    "p_value_SKA_MV":"mean"
}).reset_index()
tabla_formateada = tabla_avg_pvalue_grouped.copy()
tabla_formateada["p_value_SKA_DT"] = tabla_formateada["p_value_SKA_DT"].apply(format_pvalue)
tabla_formateada["p_value_SKA_MV"] = tabla_formateada["p_value_SKA_MV"].apply(format_pvalue)
tabla_formateada["p_value_SAoT"] = tabla_formateada["p_value_SAoT"].apply(format_pvalue)
tabla_formateada["p_value_SAoT_FS"] = tabla_formateada["p_value_SAoT_FS"].apply(format_pvalue)
tabla_formateada["p_value_AoT"] = tabla_formateada["p_value_AoT"].apply(format_pvalue)

tablas_por_corpus = {}
for c in corpus:
    tablas_por_corpus[c] = tabla_formateada[tabla_formateada["corpus"] == c].drop(columns=["corpus"]).reset_index(drop=True)
tablas_por_corpus
```

```python
# Crea una tabla agrupada por modelo y corpus, promediando los valores numéricos
tabla_por_corpus = tabla_avg_pvalue_grouped[["model", "corpus", 
                                             "base_avg", "base_std",
                                             "SKA_DT_avg", "p_value_SKA_DT","SKA_DT_std",
                                             "SKA_MV_avg", "p_value_SKA_MV", "SKA_MV_std",
                                             "AoT_avg", "p_value_AoT", "AoT_std",
                                             "SAoT_avg", "p_value_SAoT", "SAoT_std",
                                             "SAoT_FS_avg", "p_value_SAoT_FS","SAoT_FS_std"]].copy()

# Multiplica base_avg y DT_avg por 100 y redondea a 1 decimal
tabla_por_corpus["Base"] = (tabla_por_corpus["base_avg"] * 100).round(1)
tabla_por_corpus["SKA_DT"] = (tabla_por_corpus["SKA_DT_avg"] * 100).round(1)
tabla_por_corpus["SKA_MV"] = (tabla_por_corpus["SKA_MV_avg"] * 100).round(1)
tabla_por_corpus["SAoT"] = (tabla_por_corpus["SAoT_avg"] * 100).round(1)
tabla_por_corpus["SAoT_FS"] = (tabla_por_corpus["SAoT_FS_avg"] * 100).round(1)
tabla_por_corpus["AoT"] = (tabla_por_corpus["AoT_avg"] * 100).round(1)

# Formatea el p_value según los criterios dados
def format_pvalue(p):
    if p < 0.001:
        return "***"
    elif p < 0.01:
        return "**"
    elif p < 0.05:
        return "*"
    else:
        return round(p, 4)

tabla_por_corpus["p_value_SKA_DT"] = tabla_por_corpus["p_value_SKA_DT"].apply(format_pvalue)
tabla_por_corpus["p_value_SKA_MV"] = tabla_por_corpus["p_value_SKA_MV"].apply(format_pvalue)
tabla_por_corpus["p_value_SAoT"] = tabla_por_corpus["p_value_SAoT"].apply(format_pvalue)
tabla_por_corpus["p_value_AoT"] = tabla_por_corpus["p_value_AoT"].apply(format_pvalue)
tabla_por_corpus["p_value_SAoT_FS"] = tabla_por_corpus["p_value_SAoT_FS"].apply(format_pvalue)

tabla_por_corpus2=tabla_por_corpus[["model", "corpus", "Base","AoT", "p_value_AoT","SAoT", "p_value_SAoT", "SAoT_FS", "p_value_SAoT_FS",
                   "SKA_MV","p_value_SKA_MV", "SKA_DT", "p_value_SKA_DT",]]
```
