```python
import pandas as pd
from scipy.stats import ttest_ind
from scipy.stats import mannwhitneyu
```

```python
ruta="../"
#models=["phi3:medium","phi3"]
models=["gemma2:2b","gemma2","llama3.2","llama3.1","phi3:medium","phi3"]#,"phi3"]
#corpus=["SICK","DIAG"]
corpus=["SICK","SNLI"]#,"Scitail","RTEGLUE"]
labels=["Entail_f1","Neutral_f1","Contra_f1"]
```

```python
new_data={"model":[],"corpus":[],"label":[],
          "base_avg":[],"base_std":[],"base_std0":[],
          "SKA_DT_avg":[],"SKA_DT_std":[],"SKA_DT_std0":[],"p_value_SKA_DT":[],
          "SKA_MV_avg":[],"p_value_SKA_MV":[],"SKA_MV_std":[],"SKA_MV_std0":[],
          "SAoT_avg":[],"p_value_SAoT":[],"SAoT_std":[],"SAoT_std0":[],
          "SAoT_FS_avg":[],"p_value_SAoT_FS":[],"SAoT_FS_std":[],"SAoT_FS_std0":[],
          "AoT_avg":[],"p_value_AoT":[],"AoT_std":[],"AoT_std0":[]}
```

```python
for m in models:
    for c in corpus:
        #for i in range(5):
        data_results1=pd.read_pickle(ruta+m+"/"+c+"/answers/cv1/results.pickle")
        data_results2=pd.read_pickle(ruta+m+"/"+c+"/answers/cv2/results.pickle")
        data_results3=pd.read_pickle(ruta+m+"/"+c+"/answers/cv3/results.pickle")
        data_results4=pd.read_pickle(ruta+m+"/"+c+"/answers/cv4/results.pickle")
        data_results5=pd.read_pickle(ruta+m+"/"+c+"/answers/cv5/results.pickle")
        data_results=pd.concat([data_results1,data_results2,data_results3,data_results4,data_results5],axis=0)
        for la in labels:
            #print(m,c,"CV",i+1)
            #data_results=pd.read_pickle(ruta+m+"/"+c+"/answers/cv"+str(i+1)+"/results.pickle")
            base=data_results[data_results["group"]=="Base"][la]
            base_avg=data_results[data_results["group"]=="Base"][la].mean()
            base_std=data_results[data_results["group"]=="Base"][la].std()
            base_std0=data_results[data_results["group"]=="Base"][la].std(ddof=0)

            gs_dt=data_results[data_results["group"]=="SKA_DT"][la]
            gs_dt_avg=data_results[data_results["group"]=="SKA_DT"][la].mean()
            gs_dt_std=data_results[data_results["group"]=="SKA_DT"][la].std()
            gs_dt_std0=data_results[data_results["group"]=="SKA_DT"][la].std(ddof=0)

            vm=data_results[data_results["group"]=="SKA_MV"][la]
            vm_avg=data_results[data_results["group"]=="SKA_MV"][la].mean()
            vm_std=data_results[data_results["group"]=="SKA_MV"][la].std()
            vm_std0=data_results[data_results["group"]=="SKA_MV"][la].std(ddof=0)

            saot=data_results[data_results["group"]=="SAoT"][la]
            saot_avg=data_results[data_results["group"]=="SAoT"][la].mean()
            saot_std=data_results[data_results["group"]=="SAoT"][la].std()
            saot_std0=data_results[data_results["group"]=="SAoT"][la].std(ddof=0)

            saot_fs=data_results[data_results["group"]=="SAoT_FS"][la]
            saot_fs_avg=data_results[data_results["group"]=="SAoT_FS"][la].mean()
            saot_fs_std=data_results[data_results["group"]=="SAoT_FS"][la].std()
            saot_fs_std0=data_results[data_results["group"]=="SAoT_FS"][la].std(ddof=0)

            aot=data_results[data_results["group"]=="AoT"][la]
            aot_avg=data_results[data_results["group"]=="AoT"][la].mean()
            aot_std=data_results[data_results["group"]=="AoT"][la].std()
            aot_std0=data_results[data_results["group"]=="AoT"][la].std(ddof=0)

            new_data["label"].append(la)
            new_data["model"].append(m)
            new_data["corpus"].append(c)
            new_data["base_avg"].append(base_avg)
            new_data["base_std"].append(base_std)
            new_data["base_std0"].append(base_std0)
            
            new_data["SKA_DT_avg"].append(gs_dt_avg)   
            new_data["p_value_SKA_DT"].append(mannwhitneyu(base, gs_dt).pvalue)
            new_data["SKA_DT_std"].append(gs_dt_std)
            new_data["SKA_DT_std0"].append(gs_dt_std0)

            new_data["SKA_MV_avg"].append(vm_avg)   
            new_data["p_value_SKA_MV"].append(mannwhitneyu(base, vm).pvalue)
            new_data["SKA_MV_std"].append(vm_std)
            new_data["SKA_MV_std0"].append(vm_std0)

            new_data["SAoT_avg"].append(aot_avg)   
            try:
                new_data["p_value_SAoT"].append(mannwhitneyu(base, aot).pvalue)
            except:
                new_data["p_value_SAoT"].append(0)
            new_data["SAoT_std"].append(aot_std)
            new_data["SAoT_std0"].append(aot_std0)

            new_data["SAoT_FS_avg"].append(saot_fs_avg)   
            try:
                new_data["p_value_SAoT_FS"].append(mannwhitneyu(base, saot_fs).pvalue)
            except:
                new_data["p_value_SAoT_FS"].append(0)
            new_data["SAoT_FS_std"].append(saot_fs_std)
            new_data["SAoT_FS_std0"].append(saot_fs_std0)

            new_data["AoT_avg"].append(aot_avg) 
            try:  
                new_data["p_value_AoT"].append(mannwhitneyu(base, aot).pvalue)
            except:
                new_data["p_value_AoT"].append(0)
            new_data["AoT_std"].append(aot_std)
            new_data["AoT_std0"].append(aot_std0)
                
```

```python
def format_pvalue(p):
    if p < 0.001:
        return "***"
    elif p < 0.01:
        return "**"
    elif p < 0.05:
        return "*"
    else:
        return round(p, 4)

tabla_formateada = temp.copy()
tabla_formateada["p_value_SKA_DT"] = temp["p_value_SKA_DT"].apply(format_pvalue)
tabla_formateada["p_value_SKA_MV"] = temp["p_value_SKA_MV"].apply(format_pvalue)
tabla_formateada["p_value_SAoT"] = temp["p_value_SAoT"].apply(format_pvalue)
tabla_formateada["p_value_SAoT_FS"] = temp["p_value_SAoT_FS"].apply(format_pvalue)
tabla_formateada["p_value_AoT"] = temp["p_value_AoT"].apply(format_pvalue)

tablas_por_corpus = {}
for la in corpus:
    tablas_por_corpus[la] = temp[temp["corpus"] == la].reset_index(drop=True) #.drop(columns=["corpus"])
tablas_por_corpus

#pd.DataFrame(tablas_por_corpus)
```

```python
# Crea una tabla agrupada por modelo y corpus, promediando los valores numéricos
tabla_por_corpus = temp[["model",  "label", "corpus",
                                             "base_avg", "base_std",
                                             "SKA_DT_avg", "p_value_SKA_DT","SKA_DT_std",
                                             "SKA_MV_avg", "p_value_SKA_MV", "SKA_MV_std",
                                             "SAoT_avg", "p_value_SAoT", "SAoT_std",
                                             "AoT_avg", "p_value_AoT", "AoT_std",
                                             "SAoT_FS_avg", "p_value_SAoT_FS","SAoT_FS_std"]].copy()

# Multiplica base_avg y DT_avg por 100 y redondea a 1 decimal
tabla_por_corpus["Base"] = (tabla_por_corpus["base_avg"] * 100).round(1)
tabla_por_corpus["SKA_DT"] = (tabla_por_corpus["SKA_DT_avg"] * 100).round(1)
tabla_por_corpus["SKA_MV"] = (tabla_por_corpus["SKA_MV_avg"] * 100).round(1)
tabla_por_corpus["SAoT"] = (tabla_por_corpus["SAoT_avg"] * 100).round(1)
tabla_por_corpus["SAoT_FS"] = (tabla_por_corpus["SAoT_FS_avg"] * 100).round(1)
tabla_por_corpus["AoT"] = (tabla_por_corpus["AoT_avg"] * 100).round(1)

# Formatea el p_value según los criterios dados
def format_pvalue(p):
    if p < 0.001:
        return "***"
    elif p < 0.01:
        return "**"
    elif p < 0.05:
        return "*"
    else:
        return round(p, 4)

tabla_por_corpus["p_value_SKA_DT"] = tabla_por_corpus.apply(lambda x: "" if x['Base']>=x['SKA_DT'] else format_pvalue(x['p_value_SKA_DT']), axis=1)
tabla_por_corpus["p_value_SKA_MV"] = tabla_por_corpus.apply(lambda x: "" if x['Base']>=x['SKA_MV'] else format_pvalue(x['p_value_SKA_MV']), axis=1)
tabla_por_corpus["p_value_SAoT"] = tabla_por_corpus.apply(lambda x: "" if x['Base']>=x['SAoT'] else format_pvalue(x['p_value_SAoT']), axis=1)
tabla_por_corpus["p_value_SAoT_FS"] = tabla_por_corpus.apply(lambda x: "" if x['Base']>=x['SAoT_FS'] else format_pvalue(x['p_value_SAoT_FS']), axis=1)
tabla_por_corpus["p_value_AoT"] = tabla_por_corpus.apply(lambda x: "" if x['Base']>=x['AoT'] else format_pvalue(x['p_value_AoT']), axis=1)

# tabla_por_corpus["p_value_dt"] = tabla_por_corpus["p_value_dt"].apply(format_pvalue)
# tabla_por_corpus["p_value_vm"] = tabla_por_corpus["p_value_vm"].apply(format_pvalue)
# tabla_por_corpus["p_value_aot"] = tabla_por_corpus["p_value_aot"].apply(format_pvalue)
# tabla_por_corpus["p_value_cot"] = tabla_por_corpus["p_value_cot"].apply(format_pvalue)
# tabla_por_corpus["p_value_aot_fs"] = tabla_por_corpus["p_value_aot_fs"].apply(format_pvalue)

tabla_por_corpus2=tabla_por_corpus[["model", "corpus","label", "Base","AoT", "p_value_AoT","SAoT", "p_value_AoT", "SAoT_FS", "p_value_SAoT_FS",
                   "SKA_MV","p_value_SKA_MV", "SKA_DT", "p_value_SKA_DT","base_std","AoT_std","SKA_DT_std","SKA_MV_std","SAoT_std","SAoT_FS_std"]]
```

```python
orden_modelos= ['gemma2', 'gemma2:2b', 'llama3.1','llama3.2','phi3:medium','phi3']
```

```python
# Ordena el DataFrame primero por 'corpus' y luego por el orden de los modelos
tabla_ordenada_por_corpus = tabla_por_corpus2.sort_values(
    by=[ 'model','corpus','label'], 
    key=lambda col: col if col.name != 'model' else col.map({model: i for i, model in enumerate(orden_modelos)})
).reset_index(drop=True)

tabla_ordenada_por_corpus
```

```python
tabla_ordenada_por_corpus["Base"] = tabla_ordenada_por_corpus["Base"].astype(str) + " ± " + (tabla_ordenada_por_corpus["base_std"]*100).round(2).astype(str)
tabla_ordenada_por_corpus["SKA_DT"] = tabla_ordenada_por_corpus["SKA_DT"].astype(str) + " ± " + (tabla_ordenada_por_corpus["SKA_DT_std"]*100).round(2).astype(str)
tabla_ordenada_por_corpus["SAoT"] = tabla_ordenada_por_corpus["SAoT"].astype(str) + " ± " + (tabla_ordenada_por_corpus["SAoT_std"]*100).round(2).astype(str)
tabla_ordenada_por_corpus["SAoT_FS"] = tabla_ordenada_por_corpus["SAoT_FS"].astype(str) + " ± " + (tabla_ordenada_por_corpus["SAoT_FS_std"]*100).round(2).astype(str)
tabla_ordenada_por_corpus["AoT"] = tabla_ordenada_por_corpus["AoT"].astype(str) + " ± " + (tabla_ordenada_por_corpus["AoT_std"]*100).round(2).astype(str)
tabla_ordenada_por_corpus["SKA_VM"] = tabla_ordenada_por_corpus["SKA_MV"].astype(str) + " ± " + (tabla_ordenada_por_corpus["SKA_MV_std"]*100).round(2).astype(str)

tabla_ordenada_por_corpus
```
